```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pp.dtypes
```

```python
promedio_facturacion_por_pais = pp.groupby('Sales holding company')['Invoice amount gross'].mean()
plt.figure(figsize=(10, 6))
plt.hist(promedio_facturacion_por_pais, bins=20, color='blue', alpha=0.7)
plt.xlabel('Promedio de Facturación')
plt.ylabel('Frecuencia')
plt.title('Distribución del Promedio de Facturación por País')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(country_counts.index, country_counts.values, color='blue', alpha=0.7)
plt.xlabel('País')
plt.ylabel('Cantidad')
plt.title('Cantidad de Registros por País')
plt.xticks(rotation=90)
plt.tight_layout()  # Ajustar el diseño
plt.show()
```

```python
clientes = []
promedios = []
for cliente, promedio in promedio_facturacion_por_cliente.items():
    clientes.append(cliente)
    promedios.append(promedio)
```

```python
data = pd.DataFrame({'Cliente': clientes, 'Promedio de Facturación': promedios})
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Cliente', y='Promedio de Facturación', data=data)
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para mayor legibilidad
plt.xlabel('Cliente')
plt.ylabel('Promedio de Facturación')
plt.title('Promedio de Facturación por Cliente')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=promedio_facturacion_por_cliente.index, y=promedio_facturacion_por_cliente.values)
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para mayor legibilidad
plt.xlabel('Agente')
plt.ylabel('Promedio de Facturación')
plt.title('Promedio de Facturación por Agente')
plt.tight_layout()
plt.show()
```

```python
promedio_facturacion_por_cliente_sorted = promedio_facturacion_por_cliente.sort_values(ascending=False)
```

```python
top_10_agentes = promedio_facturacion_por_cliente_sorted.head(10)
```

```python
top_10_agentes_paises = [country_mapping.get(cliente, cliente) for cliente in top_10_agentes.index]
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_agentes_paises, y=top_10_agentes.values)
plt.xticks(rotation=90)  # Rotar etiquetas del eje x para mayor legibilidad
plt.xlabel('País')
plt.ylabel('Promedio de Facturación')
plt.title('Top 10 Países con Mejor Promedio de Facturación por Agente')
plt.tight_layout()
plt.show()
```
